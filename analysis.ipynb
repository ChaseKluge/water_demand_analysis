# Consumption Data Exploratory Analysis

```python
# Dependencies
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('monthly_consumption_summary.csv')
```

```python
df1 = df.groupby('month')['avg_day_kgals'].max()  #df1 contains the maximum value for avg_day_kgals by month
df2 = df.groupby('month')['mgd'].max()  #df2 contains the max value of mgd by month
```

## Plotting average day consumption (kgals) based on year, month

```python
df.boxplot('avg_day_kgpcd', 'year')
df.boxplot('avg_day_kgpcd', 'month')
```

As you can see, the peak season can see a significant rise in gpcd. Additionally, volatility increases dramatically during peak. The month with the highest average day is August with nearly 800 gpcd, with a close competitor in July.

Below, we use a histogram to examine the non-standard distributions present in July and June:

```python
med_gpcd = df['avg_day_kgpcd'].median()
mean_gpcd = df['avg_day_kgpcd'].mean()
max_day = df.loc[df['mgd'] == df['mgd'].max(), ['mgd', 'year', 'month']]

print(f'The maximum day demand was recorded in  {max_day.year.max()}/{max_day.month.max()} @ {max_day.mgd.max()} million gallons')
print(f'The median gpcd is {med_gpcd} kgals.')
print(f'The mean gpcd is {mean_gpcd} kgals.')
```
