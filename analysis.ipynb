# Consumption Data Exploratory Analysis

This simple exploratory analysis reviews some descriptive statistics for the billed water consumption in the years 2019 - 2023.

```python
# Dependencies
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('monthly_consumption_summary.csv')
```

## Plotting average day consumption (kgals) based on year, month

```python
df.boxplot('avg_day_kgpcd', 'year')
df.boxplot('avg_day_kgpcd', 'month')
df.boxplot('total_mg', 'month')
```

As you can see, the peak season can see a significant rise in gpcd. Additionally, volatility increases dramatically during peak. The month with the highest average day is August with nearly 800 gpcd, with a close competitor in September.

Interestingly, the total consumption shows that July holds the record in the five-year period for the highest consumption in a month.

```python
med_gpcd = df['avg_day_kgpcd'].median()
mean_gpcd = df['avg_day_kgpcd'].mean()
max_day = df.loc[df['total_mg'] == df['total_mg'].max(), ['total_mg', 'year', 'month']]

print(f'The maximum day demand was recorded in  {max_day.year.max()}/{max_day.month.max()} @ {max_day.total_mg.max()} million gallons')
print(f'The median gpcd is {med_gpcd} kgals.')
print(f'The mean gpcd is {mean_gpcd} kgals.')
```
